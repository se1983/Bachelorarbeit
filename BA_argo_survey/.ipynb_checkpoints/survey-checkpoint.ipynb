```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


def plot(data):
    bins=[1,2,3,4,5]
    x = [data.value_counts()[x] if x in data.value_counts() else 0 for x in bins]

    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.1, .85)})
    f.set_figwidth(20)
    f.set_figheight(6.3)

    sns.boxplot(data.values, ax=ax_box)
    ax_hist.bar(bins, x)
    ax_box.set(yticks=[])

    sns.despine(ax=ax_hist)
    sns.despine(ax=ax_box, left=False)

    f.suptitle(data.name, fontsize=20)
    #plt.text(data.min()-0.7,4.5, "Gar nicht")
    #plt.text(data.max()+0.5,4.5, "Sehr")
    
for col in df:
    plot(df[col])
```
